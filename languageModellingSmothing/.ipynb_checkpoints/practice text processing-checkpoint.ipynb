```python
import re #python module for regualar expression
```

```python
regex = re.compile(r'don\'t') #\i am suppling regular expression here, r is for raw .
#regex = re.compile(r'don\'t|go|to')
```

```python
fst = "I don't go to school."
sst = regex.sub("do not",fst)
print (fst)
print (sst)
```

```python
regx = re.compile(r'\'d')   #'\' is necessary as there are 12 metacharacters. 
```

```python
fst1 = "He'd like to eat ice-cream."
sst1= regx.sub(' would',fst1)
print(fst1)
print (sst1)
```

```python
replst = [ (r'don\'t','do not'),(r'\'d', ' would'),(r'won\'t','will not')]
```

```python
def replace(text,pattern):
    for (i,j) in pattern:
        regx = re.compile(i)
        text = regx.sub(j,text)
    print (text)
```

```python
replace("I don't want to eat food. You'd love to eat it, but I won't.",replst)
print ("I don't want to eat food. You'd love to eat it, but I won't.")
```

```python
text ="aaaabfds this is aaa aas aaa."
reg = re.compile(r'a{2}')
ss = reg.sub(" this ",text)
```

```python
ss
```

```python
reg = re.compile(r'a{2,}') #',' tells at least two, or else just take the one with >=2 number (r'a{4,7}) tells from 4 times to 7 times. 
#a* tells map all the a's from 1 to infinity
st = reg.sub(" this ", text)
print (text)
print (st)
```

```python
reg = re.compile(r'a?') #this matches empty string
# * - {0,infinity}
# + - {1,infinity}
# ? - {0,1}
text = "a aa aaa aaaa aaaaa  love"
ss = reg.sub(" this ",text)
```

```python
ss
```

```python
reg  = re.compile(r"aaa?") #searchs for 3 or 2 a's
ss = reg.sub("this",text)
ss
```

```python
html="this is a html tag here <p> ti sr </p>"   
#reg = re.compile(r"<\.>")#only traces one period inside the angle braces.
reg = re.compile(r"<..?>") #for at max two characters(any) insde the "<>"   "<.{2}>
ss = reg.sub(" this ",html)
ss
```

```python
#"\w" is for [a-zA-Z0-9_]
#"\s" is for any type of white spaces \t,\r or \n
#"\d" is for digits only


reg = re.compile(r"\s")
l = "this   here line \t line \n here."
ss = reg.sub("$",l)
ss
```
